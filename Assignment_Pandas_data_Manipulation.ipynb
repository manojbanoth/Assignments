1. Some values in the the FlightNumber column are missing. These numbers are
meant to increase by 10 with each row so 10055 and 10075 need to be put in
place. Fill in these missing numbers and make the column an integer column
(instead of a float column).

```python
Flight_Numbers=[]   #Empty list created
for i in np.arange(10045.0,10086,10): #for loop using arange function with step size 10
    Flight_Numbers.append(i) # in the given range values are appended to list
Flight_Numbers     
```

```python
df['FlightNumber']=Flight_Numbers # Flightnumbers columns is replaced with Flight_Numbers
df
```

2. The From_To column would be better as two separate columns! Split each
string on the underscore delimiter _ to give a new temporary DataFrame with
the correct values. Assign the correct column names to this temporary
DataFrame

```python
df_temp=df.From_To.str.split('_', expand=True) #Temparory data frame created
df_temp
```

```python
df_temp=df_temp.rename(columns={0:'From', 1:'To'})  #renaming the coloumn names
df_temp
```

4. Delete the From_To column from df and attach the temporary DataFrame

5. In the RecentDelays column, the values have been entered into the
DataFrame as a list. We would like each first value in its own column, each second value in its own column, and so on. If there isn't an Nth value, the value
should be NaN.
Expand the Series of lists into a DataFrame named delays, rename the columns
delay_1, delay_2, etc. and replace the unwanted RecentDelays column in df
with delays.

```python
df_RD = pd.DataFrame(df['RecentDelays'].values.tolist())    # Recent delay coloums is created as data frame with multi coloums
df_RD
```

```python
df_RD=df_RD.rename(columns=lambda x: 'delay_' + str(x+1))  # Renamning all the coloums names with delay_+Interger
df_RD
```

```python
df= pd.concat([df,df_RD], axis=1)  # Attaching the dataframe of mutlicolums recentdelays 
df
```

```python
df=df.drop(columns='RecentDelays')  #Deleting the recentdelay coloumns
df
```
